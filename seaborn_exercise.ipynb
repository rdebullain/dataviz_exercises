## Seaborn Assignment

```python
# import pandas
import pandas as pd

#import numpy 
import numpy as np

# import seaborn
import seaborn as sns

# import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pokemon.csv')
```

### Task 1: Plot a regression line between attack and defense.

```python
sns.regplot(x='ATK', y='DEF', data=df)
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Attack vs Defense')
plt.show()
```

### Task 2: Plot a regression line between attack and defense while displaying a distinction between the different generations. 

```python
# Creating the plot with increased size and enhanced aesthetics
plt.figure(figsize=(10, 6))  # Increase the figure size for better visibility
unique_hue_values = df['GENERATION'].nunique()
markers = ['x'] * unique_hue_values  # Replace 'o' with your preferred marker
sns.lmplot(x='ATK', y='DEF', hue='GENERATION', data=df, height=6, aspect=1.5, markers=markers, scatter_kws={'alpha':0.6})
plt.xlabel('Attack', fontsize=12)  # Customize the x-axis label with a larger font size
plt.ylabel('Defense', fontsize=12)  # Customize the y-axis label with a larger font size
plt.title('Attack vs Defense by Code', fontsize=14)  # Add a title with a larger font size
plt.grid(True)  # Add grid lines for better readability
plt.show()  # Display the plot
```

### Task 3: Make a boxplot for each feature of dataframe (without Generation and Legendary features) in one plot.

```python
# Specify the columns to include
columns_to_include = ['HP', 'ATK', 'DEF', 'SP_ATK', 'SP_DEF', 'SPD']

# Remove the outlier in 'SPD'
df = df[df['SPD'] < 30000]

# Melt the DataFrame into a long format
df_melted = pd.melt(df[columns_to_include])

# Create a boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='variable', y='value', data=df_melted)
plt.xticks(rotation=90)  # Rotate x-axis labels for readability

plt.show()
```

### Task 4: Plot a correlation heatmap of the pokemon dataframe (not including the features Generation and Legendary).

```python
# Specify the columns to include
columns_to_include = ['HP', 'ATK', 'DEF', 'SP_ATK', 'SP_DEF', 'SPD']

# Compute the correlation matrix
corr = df[columns_to_include].corr()

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)

plt.show()
```
